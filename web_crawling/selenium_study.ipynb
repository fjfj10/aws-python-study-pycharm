```python
# 크롬 드라이버 
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

# driver = webdriver.Chrome(ChromeDriverManager().install())
driver = webdriver.Chrome(executable_path="chromedriver")

driver.get("https://www.naver.com")

# sleep 중요 랜더링 할 시간을 벌어줘야함
time.sleep(3)

seachValue = input("검색어 입력: ")

seachInput = driver.find_element(by=By.CSS_SELECTOR, value="#query")
seachInput.send_keys(seachValue)
# seachInput입력값을 네이버의 CSS_SELECTOR가 query인 element에 넣고 Keys.RETURN으로 엔터를 입력(검색한 페이지로 들어갈 수 있다)
seachInput.send_keys(Keys.RETURN)

time.sleep(2)
# 토글버튼 오른쪽 이동 클릭
driver.find_element(by=By.XPATH, value='//*[@id="lnb"]/div[1]/div/div[1]/div/div[2]/div[2]/a').click()
time.sleep(1)
# 뉴스탭 클릭
driver.find_element(by=By.XPATH, value='//*[@id="lnb"]/div[1]/div/div[1]/div/div[1]/div[7]/a').click()
time.sleep(2)

newsTitle = driver.find_elements(by=By.CSS_SELECTOR, value='.news_tit')
for title in newsTitle:
    print(title.text)

# 2, 3페이지 넘어가는 버튼 a[숫자가 바뀐다]
# f'//*[@id="main_pack"]/div[2]/div/div/a[{2}]'
# f'//*[@id="main_pack"]/div[2]/div/div/a[{3}]'

for i in range(2, 5):
    driver.find_element(by=By.XPATH, value=f'//*[@id="main_pack"]/div[2]/div/div/a[{i}]').click()
    time.sleep(2)
    # newsTitles = driver.find_elements(by=By.CSS_SELECTOR, value='.news_tit')
    # for title in newsTitles:
    #     print(title.get_attribute(name="href"))
    newsImages = driver.find_elements(by=By.CSS_SELECTOR, value=".dsc_thumb .thumb")
    for img in newsImages:
        print(img.get_attribute(name="src"))
```
